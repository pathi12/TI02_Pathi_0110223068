```python
 Membaca Data
import pandas as pd
df = pd.read_csv('/content/GroceryStoreDataSet.csv', names = ['products'], sep = ',')
df.head()#
```

```python
# Menampilkan Data Asli Produk
df.values
```

```python
# Transformasi data ke bentuk array

data = list(df["products"].apply(lambda x:x.split(',')))
data
```

```python
# Instalasi Librari
! pip install mlxtend
```

```python
# Lakukan proses dataset dan simopan dalam dataframe
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()

data_encoded = te.fit(data).transform(data)

df = pd.DataFrame(data_encoded, columns=te.columns_)
df = df.astype(int)
df
```

```python
# Jalankan Fungsi Rules
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
```
