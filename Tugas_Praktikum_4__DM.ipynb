```python
# Import libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Load data
file_path = '/content/data_pelanggan_200.csv'
df = pd.read_csv(file_path)

# Binarisasi beberapa kolom menggunakan one-hot encoding
# Sesuaikan kolom yang ingin dianalisis, contoh: Media Sosial dan Lokasi Geografis
df_encoded = pd.get_dummies(df[['Media Sosial', 'Lokasi Geografis']])

# Menjalankan algoritma apriori untuk menemukan frequent itemsets dengan min_support 0.2
frequent_itemsets = apriori(df_encoded, min_support=0.2, use_colnames=True)

# Menerapkan association rules dengan metric lift dan min_threshold 1.0
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Menampilkan hasil aturan asosiasi
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())
```

```python
# Menjalankan algoritma apriori dengan nilai min_support lebih rendah
frequent_itemsets = apriori(df_encoded, min_support=0.1, use_colnames=True)

# Menerapkan association rules dengan nilai min_threshold lift lebih rendah
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.8)

# Tampilkan semua aturan asosiasi yang dihasilkan
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
